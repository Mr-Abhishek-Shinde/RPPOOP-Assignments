# Classes

```python
class BankAccount:
    def __init__(self, name, acc_number, acc_type, balance, pin):
        self.name = name
        self.acc_number = acc_number
        self.acc_type = acc_type
        self.balance = balance
        self.pin = pin
        
    def deposit(self, amount):
        self.balance += amount
        print(f"Amount deposited: ₹{amount}")
        
    def withdraw(self, amount, pin):
        if(self.pin == pin):
            if self.balance >= amount:
                self.balance -= amount
                print(f"Amount withdrawn: ₹{amount}")
            else:
                print("Insufficient balance!")
        else:
            print("Pin incorrect!")
            
    def  get_details(self):
        print(f"Account holder name: {self.name}")
        print(f"Account Number: {self.acc_number}")
        print(f"Account type: {self.acc_type}")
        print(f"Balance: ₹{self.balance}")
        
    def get_balance(self):
        print(f"Available balance: ₹{self.balance}")
        
```

```python
class Bank:
    def __init__(self, manager_name, manager_pin):
        self.__accounts = {}
        self.__acc_number = 10000
        self.__managerName = manager_name
        self.__managerPin = manager_pin
        
    def open_account(self, name, acc_type, pin, balance):
        self.__acc_number += 1
        new_account = BankAccount(name, self.__acc_number, acc_type, balance, pin)
        self.__accounts[self.__acc_number] = new_account
        print("Account opened successfully!")
        print(f"Your account number is: {self.__acc_number}")
        
    def close_account(self, acc_number, pin):
        if acc_number in self.__accounts:
            account = self.__accounts[acc_number]
            if(account.pin == pin):
                del self.__accounts[acc_number]
                print("Account closed successfully!")
            else:
                print("Pin incorrect!")    
        else:
            print("Account not found!")
            print("Recheck the account number entered.")
            
    def get_account(self, acc_number, pin):
        if acc_number in self.__accounts:
            account = self.__accounts[acc_number]
            if(account.pin == pin):
                return account
            else:
                print("Pin incorrect!")  
                return -1
        else:
            print("Account not found!")
            print("Recheck the account number entered.")
            return -1
            
    def transfer_funds(self, acc_num_sender, acc_num_reciever, pin_sender, amount):
        if (acc_num_sender in self.__accounts):
            if (acc_num_reciever in self.__accounts):
                account_sender = self.__accounts[acc_num_sender]
                if(account_sender.pin == pin_sender):
                    if (account_sender.balance >= amount):
                        account_reciever = self.__accounts[acc_num_reciever]
                        account_sender.balance -= amount
                        account_reciever.balance += amount
                        print(f"₹{amount} transferred to {account_reciever.name}'s account successfully.")
                        return
                    else:
                        print("Insufficient amount!")
                else:
                    print("Pin incorrect!")
            else:
                print("Reciever's account not found!")
                print("Recheck the entered account number of reciever.") 
                return
        else:
            print("Sender's account not found!")
            print("Recheck the entered account number of sender.")
            return
        
    def get_all_accounts(self, name, pin):
        if(self.__managerName == name and self.__managerPin == pin):
            print("Details of all the accounts in the bank:\n")
            for acc_num in self.__accounts:
                self.__accounts[acc_num].get_details()
                print()
        else:
            print("Access denied!")
            
    def get_total_deposits(self, name, pin):
        if(self.__managerName == name and self.__managerPin == pin):
            total_deposits = 0
            for acc_num in self.__accounts:
                total_deposits += self.__accounts[acc_num].balance
            print(f"Total deposit in the bank is: ₹{total_deposits}")
        else:
            print("Access denied!")
```

## Code to make this menu driven

```python
def displayLoginMenu():
    print("Are you manager/user?")
    print("1. Manager")
    print("2. User")
    print("3. Exit")
```

```python
def displayManagerMenu():
    print("Choose the action you want to perfrom:")
    print("1. Get details of all bank accounts")
    print("2. Get total amount deposited in the bank")
    print("3. Exit")
```

```python
def displayUserMenu():
    print("Choose the action you want to perfrom:")
    print("1. Open new account")
    print("2. Withdraw amount")
    print("3. Deposit amount")
    print("4. View available balance")
    print("5. View account details")
    print("6. Transfer funds") #transfer money from one acc to other acc in same bank
    print("7. Close account")
    print("8. Exit")
```

```python
def serve_manager():
    while (True):
        displayManagerMenu()
    
        choice = int(input("Enter your choice: "))
        print()
    
        if(choice == 1):
            manager_name = input("Enter you name: ")
            manager_pin = int(input("Enter your pin: "))
            bank.get_all_accounts(manager_name, manager_pin)
        
        elif(choice == 2):
            manager_name = input("Enter you name: ")
            manager_pin = int(input("Enter your pin: "))
            bank.get_total_deposits(manager_name, manager_pin)

        elif(choice == 3):
            print("Exiting...")
            print("Thank you sir.")
            return
        
        else:
            print("Invalid choice!")
        print()
```

```python
def serve_user():
    while (True):
        displayUserMenu()
    
        choice = int(input("Enter your choice: "))
        print()
    
        if(choice == 1):
            name = input("Enter you name: ")
            acc_type = input("Enter the account type: ")
            pin = int(input("Enter a pin to secure your account: "))
            initial_amount = int(input("Enter the amount you want to deposit: "))
            bank.open_account(name, acc_type, pin, initial_amount)
        
        elif(choice == 2):
            acc_num = int(input("Enter your account number: "))
            pin = int(input("Enter your pin: "))
            if(bank.get_account(acc_num, pin) == -1):
                continue
            else:
                acc = bank.get_account(acc_num, pin)
                amount = int(input("Enter the amount you want to withdraw: "))
                try:
                    acc.withdraw(amount, pin)
                except:
                    continue
            
        elif(choice == 3):
            acc_num = int(input("Enter your account number: "))
            pin = int(input("Enter your pin: "))
            if(bank.get_account(acc_num, pin) == -1):
                continue
            else:
                acc = bank.get_account(acc_num, pin)
                amount = int(input("Enter the amount you want to deposit: "))
                acc.deposit(amount)
           
        elif(choice == 4):
            acc_num = int(input("Enter your account number: "))
            pin = int(input("Enter your pin: "))
            if(bank.get_account(acc_num, pin) == -1):
                continue
            else:
                acc = bank.get_account(acc_num, pin)
                acc.get_balance()
        
        elif(choice == 5):
            acc_num = int(input("Enter your account number: "))
            pin = int(input("Enter your pin: "))
            if(bank.get_account(acc_num, pin) == -1):
                continue
            else:
                acc = bank.get_account(acc_num, pin)
                acc.get_details()
        
        elif(choice == 6):
            acc_num_sender = int(input("Enter your account number: "))
            pin_sender = int(input("Enter your pin: "))
            acc_num_reciever = int(input("Enter the account number of reciever: "))
            amount = int(input("Enter the amount you want to transfer: "))
            bank.transfer_funds(acc_num_sender, acc_num_reciever, pin_sender, amount)
        
        elif(choice == 7):
            acc_num = int(input("Enter your account number: "))
            pin = int(input("Enter your pin: "))
            if(bank.get_account(acc_num, pin) == -1):
                continue
            else:
                acc = bank.get_account(acc_num, pin)
            bank.close_account(acc_num, pin)

        elif(choice == 8):
            print("Exiting...")
            print("Thank you for banking with us.")
            return
        
        else:
            print("Invalid choice!")
        print()
```

```python
while (True):
    displayLoginMenu()
    
    choice = int(input("Enter your choice: "))
    print()
    
    if(choice == 1):
        serve_manager()
        break
        
    elif(choice == 2):
        serve_user()
        break

    elif(choice == 3):
        print("Exiting...")
        break
        
    else:
        print("Invalid choice!")
```
